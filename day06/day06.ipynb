```python
from collections import defaultdict
import sys
sys.setrecursionlimit(150000)

movement_map = {
    '>': (0, 1),
    '<': (0, -1),
    'v': (1, 0),
    '^': (-1, 0)
}

rotation_map = {
    '^': '>',
    '>': 'v',
    'v': '<',
    '<': '^'
}

obstacle = '#'

def read_file_into_2d():
    with open('day06_input.txt', 'r') as file:
        grid = [list(line.strip()) for line in file]
    return grid

def find_start(grid):
    for row in range(len(grid)):
        for col in range(len(grid[0])):
            if grid[row][col] in ['^', '>', '<', 'v']:
                return (row, col)
    return None

def track_movement(row, col, grid, curr):
    move_row, move_col = movement_map[curr]
    next_row, next_col = row + move_row, col + move_col
    
    if next_row < 0 or next_col < 0 or next_row >= len(grid) or next_col >= len(grid[0]):
        grid[row][col] = 'X'
        return 1
        
    if grid[next_row][next_col] == '.' or grid[next_row][next_col] == 'X':
        grid[row][col] = 'X'
        val = 1 if grid[next_row][next_col] == '.' else 0
        return val + track_movement(next_row, next_col, grid, curr)
        
    next_curr = rotation_map[curr]
    
    return track_movement(row, col, grid, next_curr)

def part1():
    grid = read_file_into_2d()
    rows, cols = len(grid), len(grid[0])

    curr_row, curr_col = find_start(grid)
    
    for row in range(rows):
        for col in range(cols):
            if (row, col) == (curr_row, curr_col):
                results = track_movement(row, col, grid, grid[row][col])

                print(results)
                return       

def track_movement_p2(row, col, grid, curr, visited):
    if (row, col, curr) in visited:
        return True
    
    visited.add((row, col, curr))
    
    move_row, move_col = movement_map[curr]
    next_row, next_col = row + move_row, col + move_col
    
    if next_row < 0 or next_col < 0 or next_row >= len(grid) or next_col >= len(grid[0]):
        return False

    if grid[next_row][next_col] != '#':
        return track_movement_p2(next_row, next_col, grid, curr, visited)

    next_curr = rotation_map[curr]
    return track_movement_p2(row, col, grid, next_curr, visited)


def part2():
    grid = read_file_into_2d()
    rows, cols = len(grid), len(grid[0])

    curr_row, curr_col = find_start(grid)
    total = 0
    for row in range(rows):
        for col in range(cols):
            if (row, col) == (curr_row, curr_col) or grid[row][col] == '#':
                continue
                
            grid[row][col] = '#'
            
            if track_movement_p2(curr_row, curr_col, grid, grid[curr_row][curr_col], set()):
                total += 1
                
            grid[row][col] = '.'
    print(total)
```

```python
part1()
part2()
```
